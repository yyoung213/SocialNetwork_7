```python
import pandas as pd
import glob # 여러 파일을 한 번에 불러오기 위한 라이브러리
import os
```

```python


# 1. CSV 파일들이 저장된 폴더 경로
# (만약 39개 파일을 'crawled_data'라는 폴더에 넣었다면)
# folder_path = 'crawled_data/'
# (만약 39개 파일을 그냥 현재 위치에 다 올렸다면)
folder_path = 'C:\\socialnetwork\\SocialNetwork_7\\raw_data' 

# 2. 해당 폴더 내의 모든 .csv 파일 경로를 리스트로 가져오기
# (os.path.join은 'crawled_data/*.csv' 처럼 경로를 합쳐줍니다)
file_pattern = os.path.join(folder_path, '*.csv')
file_list = glob.glob(file_pattern)

print(file_list)
if not file_list:
    print(f"오류: '{folder_path}' 폴더에서 CSV 파일을 찾을 수 없습니다.")
    print("39개 파일을 업로드했는지, 경로가 맞는지 확인해주세요.")
else:
    print(f"총 {len(file_list)}개의 CSV 파일을 찾았습니다. 통합을 시작합니다...")
    
    all_df_list = []
    
    # 3. 각 파일을 순회하며 읽어오고 리스트에 추가
    for file_path in file_list:
        try:
            df = pd.read_csv(file_path)
            all_df_list.append(df)
            print(f" - '{file_path}' ({len(df)} 행) 로드 완료")
        except Exception as e:
            print(f" - '{file_path}' 로드 중 오류 발생: {e}")

    # 4. 리스트에 담긴 모든 DataFrame을 하나로 합치기
    if all_df_list:
        combined_df = pd.concat(all_df_list, ignore_index=True)
        
        # 5. 합쳐진 파일을 새 CSV로 저장
        output_filename = 'csv_merged.csv'
        combined_df.to_csv(output_filename, index=False, encoding='utf-8-sig') # (utf-8-sig는 Excel에서 한글 깨짐 방지)
        
        print("\n--- 통합 완료 ---")
        print(f"총 {len(combined_df)}개의 채용 공고가 '{output_filename}' 파일로 저장되었습니다.")
        print("\n[통합 데이터 정보]")
        combined_df.info()
        print("\n[통합 데이터 샘플]")
        print(combined_df.head())
    else:
        print("오류: 파일을 읽는 데 모두 실패했습니다.")
```
