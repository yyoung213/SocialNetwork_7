```python
skill_keys = list(SKILL_DICT.keys())
print(f"Step 2/5: 스킬 사전 정의 완료 (총 {len(skill_keys)}개 키)")
```

```python
# --- 4. 스킬 정규화 함수 정의 ---
def extract_skills(text):
        if not isinstance(text, str):
            return []
        
        text_lower = text.lower() # 전체 텍스트를 소문자로 변경
        found_skills = set() # 한 공고 내 중복 스킬 제거를 위해 set 사용
        
        for skill_key in skill_keys:
            # \b는 단어 경계(word boundary)를 의미합니다.
            if re.search(r'\b' + re.escape(skill_key) + r'\b', text_lower):
                found_skills.add(SKILL_DICT[skill_key]) # 정규화된 이름(Value)을 set에 추가
                
        return list(found_skills) # set을 list로 변환하여 반환
```

```python
# --- 5. 함수 적용 및 Bipartite 변환 ---
print("Step 3/5: 공고에 스킬 추출 함수 적용 중... (시간이 걸릴 수 있습니다)")
df['normalized_skills'] = df['skills_text'].apply(extract_skills)
```

```python
bipartite_df = df.explode('normalized_skills')
```

```python
bipartite_df = bipartite_df.dropna(subset=['normalized_skills'])
print("Step 4/5: 'explode' 변환 완료. Bipartite Edge List 생성 중...")
```

```python
# 사용자가 요청한 [기업명, 스킬] 2개 컬럼만 선택
final_edges_df = bipartite_df[['기업명', 'normalized_skills']]
    
    # 컬럼명 변경 (normalized_skills -> Skill)
final_edges_df = final_edges_df.rename(columns={'normalized_skills': 'Skill'})
    
    # (선택) 기업-스킬 조합이 중복되는 경우 제거
final_edges_df = final_edges_df.drop_duplicates().reset_index(drop=True)

    # 최종 CSV 파일로 저장
output_filename = 'bipartite_skill_edges.csv'
final_edges_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
```
