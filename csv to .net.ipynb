```python
import pandas as pd
import networkx as nx
```

```python
# 1. CSV 파일을 pandas DataFrame으로 읽기
df = pd.read_csv("C:\\socialnetwork\\SocialNetwork_7\\bipartite_skill_long.csv")

# 2. DataFrame에서 그래프 객체 생성
# (만약 가중치(Weight)가 없다면 'edge_attr=True' 옵션 제외)
G = nx.from_pandas_edgelist(
    df, 
    source='기업명', 
    target='Skill', 
)

# 3. Pajek .net 파일로 저장
# networkx가 노드 이름을 정수로 자동 매핑하여 저장합니다.
nx.write_pajek(G, "pajek_data.net")

print("pajek_data.net 파일이 성공적으로 생성되었습니다.")
```
